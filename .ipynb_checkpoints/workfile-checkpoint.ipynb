```python
import csv, geopy

from geopy.geocoders import GoogleV3
from geopy.exc import GeocoderTimedOut

with open('./workfile/master.csv', mode='r', encoding='utf-8-sig') as infile:
    reader = csv.DictReader(infile)
    dict_list = []
    out_list = []
    
    geolocator = GoogleV3(api_key="AIzaSyB0SfctYbf8P7kOOqTlo9gMJFKhlLKJBzA", domain="maps.google.com.my")
    
    line_num = 1
    
    for line in reader:
        
        # Append current line to dictionary for backup
        dict_list.append(line)
        
        print("processing line", line_num)
        line_num = line_num + 1
        
        if line["Location"] == "":
            line["latitude"] = 0
            line["longitude"] = 0

        else:
            out = geolocator.geocode(line["Location"], timeout=10, components=[('country', 'Malaysia'), ('administrative_area', line["Region"])])

            if out == None:
                line["latitude"] = 0
                line["longitude"] = 0
            else:
                line["latitude"] = out.latitude
                line["longitude"] = out.longitude
        
        out_list.append(line)
                
```
